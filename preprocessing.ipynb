```python
def draw_hypnogram(y, y_stages, ax, name=None, ):
    assert len(y) == len(y_stages)
    ax2 = ax.twinx()
    ax.step(y, 'b-')
    ax2.step(y_stages, 'g-')
    ax2.invert_yaxis()
    ax.set_xlabel('time [s]')
    ax.set_ylabel('Voltage [V]', color='b')
    ax2.set_ylabel('Sleep Stage', color='g')
    ax.ticklabel_format(style='sci', axis='y', scilimits=(0, 0))
    labels = [m.stages_names[item] if item in m.stages_names else item for item in ax2.get_yticks()]
    ax2.set_yticklabels(labels)
    plt.legend(['ECG', 'stage'])
    if name is not None:
        ax.set_title(f"ECG of {name}")
```

```python
def where_only_one_spike(data, max= 0.0033):
    hits = []
    for i in range(1, len(data) - 1):
        prev = abs(data[i - 1])
        cur = abs(data[i])
        nextt = abs(data[i + 1])
        if cur == max and prev < max and nextt < max:
            hits.append(i)

    return hits
```

```python
possible_interpolation = where_only_one_spike(all_data.flatten())
len(possible_interpolation)
```
