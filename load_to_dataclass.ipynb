```python
import numpy as np
import wfdb.processing
import mne
import MyEDFImports as m
from dataclasses import dataclass
from typing import Optional
```

```python
@dataclass
class labeled_window:
    patient: str
    stage: int
    data: np.ndarray
    index: Optional[int] = None
    detected_peaks: Optional[list] =  None
```

```python
#check all sampling freq
files = m.get_edf_filenames()
sfreqs = []
for file in files:
    edf_dir = m.edf_dir
    edf = mne.io.read_raw_edf(edf_dir + '//' + file)
    sfreq = edf.info['sfreq']
    sfreqs.append(sfreq)
```

```python
sfreqs
```

```python
sampling_freq = 500
cur_ind = 0
for file in files:
    print(file)
    data = m.load_data_one_file(file)
    labels = m.load_labels_one_file(file)
    assert len(data) == len(labels)
    peaks_entire_sleep = wfdb.processing.xqrs_detect(data.flatten(), fs=sampling_freq)
    window_peaks = []
    for i in range(len(labels)):
        start_ind = i * 20 * sampling_freq
        end_ind = (i + 1) * 20 * sampling_freq
        window_peaks.append(list(filter(lambda x: start_ind <= x < end_ind, peaks_entire_sleep)))
    all_windows = []
    for i, (d, l, p) in enumerate(zip(data, labels, window_peaks)):
        new_wind = labeled_window(patient=file,
                                  data=d,
                                  stage=l,
                                  index=i + cur_ind,
                                  detected_peaks=p)
        all_windows.append(new_wind)

```

```python
edf_filenames = m.get_edf_filenames()
sfreq = 500
patients = {}
for name in edf_filenames[:1]:
    print(name)
    data = m.load_data_one_file(name)
    labels = m.load_labels_one_file(name)
    all_peaks = wfdb.processing.xqrs_detect(data.flatten(), fs=sfreq)
    window_peaks = []
    cur_win = []
    cur_win_lim = 0
    for peak in all_peaks:
        if peak < cur_win_lim:
            cur_win.append(peak)
        else:
            window_peaks.append(cur_win)
            cur_win = []
            cur_win_lim += 20*500
    assert len(window_peaks) == len(data)
    patients[name] = zip(data,labels,window_peaks)
```
